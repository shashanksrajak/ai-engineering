## Training Pipeline in PyTorch
this shows how a training pipeline is built using pytorch. this consider a single neuron.

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv(
    'https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
```

```python
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2, random_state=42)
```

```python
X_train.head()
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
type(X_train)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

So far its good and now we will build neural network, but this needs tensors as input instead of numpy arrays.

```python
X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
X_train_tensor.shape
```

```python
y_train_tensor.shape
```

### Define the nn model

```python
class MySimpleNN:
    def __init__(self, X):
        """
        X: input tensor
        """
        
        # we are using one single neuron, 30 features so 30 weights and 1 bias
        self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
        self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)
        
    def forward(self, X):
        z = torch.matmul(X, self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred
    
    def loss(self, y_pred, y):
        epsilon = 1e-7
        y_pred = torch.clamp(y_pred, epsilon, 1-epsilon)
        
        # binary cross entropy loss for classification
        loss = -(y * torch.log(y_pred) + (1-y) * torch.log(1-y_pred)).mean()
        
        return loss
```

```python
learning_rate = 0.1
epochs = 25 # how many runs over the data
```

```python
# create model
model = MySimpleNN(X_train_tensor)
```

```python
model.weights.shape
```

```python
# define training loop
for epoch in range(epochs):
    # forward pass
    y_pred = model.forward(X_train_tensor)
    
    # loss calculate
    loss = model.loss(y_pred, y_train_tensor)
        
    # backward pass (backprop)
    loss.backward()
    
    # update parameters w and b
    with torch.no_grad():
        # !!!caution - below line will break
        # model.bias = model.bias - learning_rate * model.bias.grad

        model.weights -= learning_rate * model.weights.grad
        model.bias -=  learning_rate * model.bias.grad
        
    # set gradients to zero because they accumulate 
    model.weights.grad.zero_()
    model.bias.grad.zero_()
    
    print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```

```python
model.bias
```

#### Evaluation

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    
    # y_pred will be 0-1 but we need to have 0/1 so we decide a threshold and then convert
    threshold = 0.5
    
    y_pred = (y_pred > threshold).float()
    
    accuracy = (y_pred == y_test_tensor).float().mean()
    
    print(f"Accuracy {accuracy.item()}")
```
